```python
death_df = pd.read_csv(f'{data_root}JHU_COVID_Deaths.csv')
death_df.head()
```

```python
surv_df = pd.read_csv(f'{data_root}overall-county-smoothed.csv')
surv_df.head()
```

```python
mobi_df = pd.read_csv(f'{data_root}Global_Mobility_Report.csv')
mobi_df.head()
```

```python
mobi_df.shape
```

```python
mobi_df = mobi_df[mobi_df.country_region_code=='US']
mobi_df.shape
```

```python
mobi_df.head()
```

^ as you can see there are many missing values

```python
mobi_df.census_fips_code.unique()
```

```python
# remove records with no fips value
mobi_df = mobi_df[~mobi_df.census_fips_code.isna()]
mobi_df['census_fips_code'] = mobi_df['census_fips_code'].astype('int64')
mobi_df.shape
```

```python
mobi_df['Key'] = mobi_df['date'].astype(str)+'_'+mobi_df['census_fips_code'].astype(str)
mobi_df.Key.head()
```

### The survey date starts from12-April-2020

```python
# Let's remove the records before that date
death_df = death_df[death_df.Date>'2020-04-12']
death_df.shape
```

```python
death_df['Date'] = pd.to_datetime(death_df['Date'])
death_df.Date.head()
```

```python
death_df['Key'] = death_df['Date'].astype(str)+'_'+death_df['FIPS'].astype(str)
death_df.Key.head()
```

```python
surv_df['Key'] = surv_df['date'].astype(str)+'_'+surv_df['fips'].astype(str)
surv_df.Key.head()
```

```python
df = surv_df.merge(death_df, on='Key', how='left')
df.shape
```

```python
surv_df.shape, death_df.shape
```

```python
df.DeathRate1000.describe()
```

```python
df_master = df.merge(mobi_df, on='Key', how='left')
df_master.shape
```

```python
df_master.to_csv(f'{data_root}/merged_sruvery_death_mobility_data.csv')
```
